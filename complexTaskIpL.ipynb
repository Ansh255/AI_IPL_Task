```python
#BruteForce Approach ? First Comes to mind

condition_first  = dataframe_of_excel['Bat_Second_15_ov_wkts_lost'] == 4  
#the number of wickets lost by the team batting second in the first 15 overs is equal to 4

condition_second = dataframe_of_excel['Bat_Second_15_ov_score'] == dataframe_of_excel['Bat_First_Runs_Scored'] - 32
#score of the team batting second in the first 15 overs is equal to the runs scored by the team batting first, minus 32 runs

condition_third = dataframe_of_excel['Winner'] == dataframe_of_excel['Team_Batting_First']
#matches where the team that batted first ended up winning the game

condition_check = condition_first & condition_second & condition_third
#this condition check is merger of the all the three conditions to get such row to tuple which satisfy all three conditions

specified_match = dataframe_of_excel[condition_check]
specified_match
```

```python
dataframe_result_complex = specified_match[['Match_Number','Team_Batting_Second','Team_Batting_First','Date']]
```

```python
#Optimised Approach ?
matching_rows = []

# Iterate through the rows of the DataFrame
for index, row in dataframe_of_excel.iterrows():
    
    #print(index)
    #print(row)
    if (row['Bat_Second_15_ov_wkts_lost'] == 4 and
        row['Bat_Second_15_ov_score'] == row['Bat_First_Runs_Scored'] - 32 and
        row['Winner'] == row['Team_Batting_First']):
        matching_rows.append(row)

# Create a new DataFrame with the matching rows
result_df = pd.DataFrame(matching_rows)
result_df
```

```python
# Lambda Function
result_df = dataframe_of_excel[dataframe_of_excel.apply(lambda row: (row['Bat_Second_15_ov_wkts_lost'] == 4 and
                                      row['Bat_Second_15_ov_score'] == row['Bat_First_Runs_Scored'] - 32 and
                                      row['Winner'] == row['Team_Batting_First']), axis=1)]
result_df
```

```python
#Task 2 Starts here !

#Breakdown of Question?
#condition1 ?  Bat First : lost 2 wicket at 15 over -----------> target tuple? : Bat_First_15_ov_wkts_lost
#Condition2 ?  but score <25 runs in last 5 over  -----------> target tuple? : Bat_First_20_ov_score - Bat_First_15_ov_score < 25

```

```python
# Using LOOPS
# Create an empty list to store the matching rows
matching_rows = []

# Iterate through the rows of the DataFrame
for index, row in dataframe_of_excel.iterrows():
    # Check if all conditions are met
    if (row['Bat_First_15_ov_wkts_lost'] == 2 and
        row['Bat_First_20_ov_score'] - row['Bat_First_15_ov_score'] < 25):
        # Append the matching row to the list
        matching_rows.append([row['Match_Number'], row['Team_Batting_Second'], row['Team_Batting_First'], row['Date']])

# Create a new DataFrame with the matching rows
result_dataframe = pd.DataFrame(matching_rows, columns=['Match_Number', 'Team_Batting_Second', 'Team_Batting_First', 'Date'])
result_dataframe
```
